```python
df3 = df.dropna(axis=0)
df3.shape
```

```python
df3.isna().sum()
```

```python
X = df3.loc[:, df.columns != 'Global_Sales']
y = df3['Global_Sales']
random_state=59

X_train,X_other,y_train,y_other = train_test_split(X,y,train_size = 0.6,random_state = random_state)
print('training set: ',X_train.shape, y_train.shape)
print(X_other.shape, y_other.shape)

X_val, X_test, y_val, y_test = train_test_split(X_other,y_other,train_size = 0.5,random_state = random_state)
print('validation set:',X_val.shape, y_val.shape)
print('test set:',X_test.shape, y_test.shape)

print(X_train.head())
print(y_train.head())
```

```python
onehot_ftrs = ['Name','Platform', 'Genre', 'Publisher', 'Developer', 'Rating']
standard_scale_ftrs = ['Year_of_Release', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count']
preprocessor = ColumnTransformer(
    transformers =[
        ('oneHot', OneHotEncoder(sparse=False,handle_unknown='ignore'), onehot_ftrs),
        ('std', StandardScaler(), standard_scale_ftrs)])
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

X_train_prep = pipeline.fit_transform(X_train)
X_val_prep = pipeline.transform(X_val)
X_test_prep = pipeline.transform(X_test)

print(X_train.shape)
print(X_train_prep.shape)
print(X_val.shape)
print(X_val_prep.shape)
print(X_test.shape)
print(X_test_prep.shape)
print(X_train_prep)
```
